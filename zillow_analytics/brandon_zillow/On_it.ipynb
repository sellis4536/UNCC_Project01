```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import csv 
csvfile = "Zsample10.csv"
zillow_df = pd.read_csv(csvfile)
```

```python
zillow_df.head()
```

```python
countyLA = zillow_df.loc[zillow_df["fips"] == 6037, :]
countyOC = zillow_df.loc[zillow_df["fips"] == 6059, :]
countyVC = zillow_df.loc[zillow_df["fips"] == 6111, :]
```

```python
losAngeles = countyLA.set_index(["fips"])
losAngeles.rename({6037: "Los Angeles"})
orangeCounty = countyOC.set_index(["fips"])
orangeCounty.rename({6059: "Orange County"})
venturaCounty = countyVC.set_index(["fips"])
venturaCounty.rename({6111: "Ventura County"})
```

# Average Square Feet

```python
losAngelesAverageFt = countyLA["calc_sqft"].mean()
orangeCountyAverageFt = countyOC["calc_sqft"].mean()
venturaCountyAverageFt = countyVC["calc_sqft"].mean()
```

# Average Beds

```python
losAngelesAvgBed = losAngeles["beds"].mean()
orangeCountyAvgBed = orangeCounty["beds"].mean()
venturaCountyAvgBed = venturaCounty["beds"].mean()
```
